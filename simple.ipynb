```python
# love connected <3
import random
import numpy as np
import random
from datetime import datetime
def simple_oracle(Mr, Mrs, Mr_to_Mrs_Right):
	if(Mr_to_Mrs_Right[Mr] != Mrs):
		return False;
	else:
		return True;
def sequential_algorithm(decision_map):
	my_picks = []
	for i in range(100):
		for j in range(100):
			if(decision_map[i][j] == 0):
				my_picks.append([i, j])
			if(len(my_picks) == 10):
				return my_picks
def event(algorithm):
	random.seed(datetime.now())
	Mr_to_Mrs_Right = [i for i in range(100)]
	random.shuffle(Mr_to_Mrs_Right)
	# 0 for "unknown", 1 for "perfect match", -1 for "not perfect match"
	decision_map = [[0 for i in range(100)] for j in range(100)]
	
	rounds, perfect_pick_found = 0, 0
	while(perfect_pick_found < 10):
		next_picks = algorithm(decision_map)
		rounds += 1
		for i in range(10):
			if(simple_oracle(next_picks[i][0], next_picks[i][1], Mr_to_Mrs_Right)):
				decision_map[next_picks[i][0]][next_picks[i][1]] = 1
				perfect_pick_found += 1
			else:
				decision_map[next_picks[i][0]][next_picks[i][1]] = -1
	return rounds
def random_experiment(algorithm, iteration):
	sample_space, mean = [], 0
	for i in range(iteration):
		n = event(algorithm)
		print("Iteration %d: outcome is %d times" %(i, n))
		sample_space.append(n)
		mean += n
	return mean / iteration, sample_space

mean, history = random_experiment(sequential_algorithm, 10000)
print(mean, history)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl
import statistics
dev = statistics.stdev(history)
mpl.style.use('seaborn')
plt.hist(history, color = '#fccf0d', edgecolor='#2a2927', linewidth=4.0)
plt.title("sequential_algorithm")
plt.text(mean, 1200, r'$\mu=%f,\ \sigma=%s$'%(mean, dev))
plt.xlabel("Picking Rounds")
plt.ylabel("Frequency")
# plt.gca().invert_xaxis()
# plt.gca().invert_yaxis()
fig = plt.gcf()
plt.savefig('sequential_algorithm.png')
```
